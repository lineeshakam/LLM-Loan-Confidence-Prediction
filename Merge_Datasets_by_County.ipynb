```python
import pandas as pd
model = "gpt-oss-20b"

# Load both datasets
df1 = pd.read_csv('us_county_llm_estimates (2) (1).csv')  # Your first dataset
df2 = pd.read_csv('cleaned_avg_credit_score_2020_fips_avg.csv')  # Your second dataset

# Merge datasets on the matching columns (FIPS in df1 and county_fips in df2)
merged_df = pd.merge(df1, df2, left_on='county_fips', right_on='FIPS', how='inner')

# Export the merged data to a new CSV
merged_df.to_csv(f'merged_data_{model}.csv', index=False)
files.download(f'merged_data_{model}.csv')
```
