```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#GPT-oss-20b

```python
gptoss20b = pd.read_csv('merged_data_gpt-oss-20b.csv')
gptoss20b.head()
```

```python
gptoss20b.info()
```

```python
sumNull = gptoss20b.isnull().sum()
percentage = sumNull/ gptoss20b.shape[0] * 100
print(percentage)
```

```python
predictedCreditNa = gptoss20b[gptoss20b['predicted_credit_score'].isna()]
predictedCreditNa
```

```python
gptoss20b['predicted_credit_score'].hist(bins=10) # 'bins' controls the number of intervals
plt.title('Histogram of Predicted Credit Score')
plt.xlabel('Predicted Credit Scores')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
gptoss20b['credit_score'].hist(bins=10) # 'bins' controls the number of intervals
plt.title('Histogram of Actual Credit Score')
plt.xlabel('Actual Credit Scores')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
gptoss20b['Difference'].hist(bins=10) # 'bins' controls the number of intervals
plt.title('Histogram of Difference in Credit Score')
plt.xlabel('Difference in Credit Scores')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Plot both histograms on the same axes
plt.hist(gptoss20b['credit_score'], bins=10, alpha=0.6, label='Actual Credit Score', color='skyblue', edgecolor='black')
plt.hist(gptoss20b['predicted_credit_score'], bins=10, alpha=0.6, label='Predicted Credit Score', color='purple', edgecolor='black')

# Add labels and legend
plt.title('Actual vs Predicted Credit Score Distribution gpt-oss-20b')
plt.xlabel('Credit Scores')
plt.ylabel('Frequency')
plt.legend()
plt.grid(False)
plt.show()
```
